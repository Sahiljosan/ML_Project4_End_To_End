### 3.7) Adding columns for "Total Score" and "Average"

````python
reading_full = df[df['reading_score'] <= 20]['average'].count()
writing_full = df[df['writing_score'] <= 20]['average'].count()
math_full = df[df['math_score'] <= 20]['average'].count()

print(f"Number of students with less then 20 marks in maths : {math_full}")
print(f"Number of students with less then 20 marks in Writing : {writing_full}")
print(f"Number of students with less then 20 marks in reading : {reading_full}")
````
